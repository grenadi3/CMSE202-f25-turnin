## Part 3: Logistic Regression Analysis (19 points)

**&#9989; Task 3.1:** Preprocess your data appropriately for logistic regression:

An important part of the preparing the data for the ML part is scaling and encoding. We have not covered this in class, so I am not asking you to do it. The code below takes care of this run it and make sure you don't get any errors.

````python
# Run this code without modifying it, apart from the name of the dataset

from sklearn.preprocessing import OneHotEncoder, StandardScaler

# Preprocess data for logistic regression
# Handle categorical variables (one-hot encoding for 'origin')
encoder = OneHotEncoder(sparse_output=False, drop='first')
origin_encoded = encoder.fit_transform(X_train[['origin']])
origin_cols = [f'origin_{i+2}' for i in range(origin_encoded.shape[1])]
origin_train_df = pd.DataFrame(origin_encoded, columns=origin_cols, index=X_train.index)

# Standardize numerical features
scaler = StandardScaler()
numerical_cols = ['cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year']
X_train_num = pd.DataFrame(
    scaler.fit_transform(X_train[numerical_cols]),
    columns=numerical_cols,
    index=X_train.index
)

# Combine numerical and encoded categorical features
X_train_processed = pd.concat([X_train_num, origin_train_df], axis=1)

# Process test data
origin_encoded_test = encoder.transform(X_test[['origin']])
origin_test_df = pd.DataFrame(origin_encoded_test, columns=origin_cols, index=X_test.index)

X_test_num = pd.DataFrame(
    scaler.transform(X_test[numerical_cols]),
    columns=numerical_cols,
    index=X_test.index
)

X_test_processed = pd.concat([X_test_num, origin_test_df], axis=1)

print("\nProcessed training data shape:", X_train_processed.shape)
print("Processed testing data shape:", X_test_processed.shape)
print("\nFirst 5 rows of processed training data:")
display(X_train_processed.head())
````

**&#9989;(7 points) Task 3.2:** Fit a logistic regression model using the `statsmodels` library:

* Use the training data to fit the model (2 points)
* Print the summary of your model results (1 point)
* Interpret the coefficients and statistical significance of features (4 points)

````python
X_train_const = sm.add_constant(X_train_processed)
logit_model = sm.Logit(y_train, X_train_const)
result = logit_model.fit()
print(result.summary())
````

* Interpret the coefficients of the logistic regression model here. What are the most important features? Why? 
How does this compare to the correlation matrix from above?

Based on the logistic regression results, weight, model year, and origin_2 are statistically significant predictors (p < 0.05) of a car having high MPG. Increased vehicle weight significantly decreases the likelihood (log-odds) of achieving high MPG, representing the strongest negative predictor in this model. Conversely, increases in model year (newer cars) and being from origin_2 (relative to origin 1) significantly increase the likelihood of having high MPG. Other features like cylinders, displacement, and acceleration did not show a statistically significant independent association with high MPG when controlling for these factors. 

The strong negative correlation between mpg and weight, and the positive correlation between mpg and model year shown in the heatmap, are consistent with these features being significant predictors in the logistic regression model. However, while features like cylinders and displacement show strong correlations with mpg, they were not significant independent predictors in the regression, likely because their influence is captured by other correlated variables like weight within the model.

**&#9989; (6 points) Task 3.3:** Evaluate the logistic regression model:

* Make predictions on the test set (1 point)
* Create a confusion matrix (1 point)
* Display the confusion matrix as a heatmap with Blues colormap (2 points)
* Calculate and print accuracy, precision, recall (2 points)

````python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score

X_test_const = sm.add_constant(X_test_processed, has_constant='add') # Explicitly add if not present

expected_columns = result.model.exog_names
X_test_const = X_test_const.reindex(columns=expected_columns, fill_value=0)
predicted_probabilities = result.predict(X_test_const)
# Convert probabilities to binary predictions (0 or 1) using a 0.5 threshold
predictions = (predicted_probabilities >= 0.5).astype(int)

cm = confusion_matrix(y_test, predictions)


# --- Display the confusion matrix as a heatmap ---
# plt.figure(figsize=(6, 4)) # Optional: Adjust figure size
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Predicted 0', 'Predicted 1'],
            yticklabels=['Actual 0', 'Actual 1'])
plt.xlabel('Predicted Label')
plt.ylabel('Actual Label')
plt.title('Confusion Matrix Heatmap')


# --- Calculate and print accuracy, precision, recall ---
reg_accuracy = accuracy_score(y_test, predictions)
# Calculate precision and recall for the positive class (label 1)
reg_precision = precision_score(y_test, predictions, pos_label=1)
reg_recall = recall_score(y_test, predictions, pos_label=1)

print(f"\nModel Evaluation Metrics:")
print(f"Accuracy: {reg_accuracy:.4f}")
print(f"Precision (for class 1): {reg_precision:.4f}")
print(f"Recall (for class 1): {reg_recall:.4f}")
````

## Part 4: Support Vector Machine Analysis (14 points)

**&#9989; (4 points) Task 4.1:** Implement an SVM classifier with RBF kernel:
- Initialize an SVM model with RBF kernel (1 point)
- Define a grid of hyperparameters to search (C and gamma) (2 points)
- Use the option `cv = 5` when fitting (This is 5-fold cross-validation to find the optimal parameters) (1 point)
- Report the best parameters found and the cross-validation score (1 point)

````python
svm_rbf = SVC(kernel='rbf')


# --- Define a grid of hyperparameters to search ---
# Define ranges for C and gamma parameters
param_grid = {
    'C': [0.1, 1, 10, 100],  # Regularization parameter
    'gamma': ['scale', 'auto'] # Kernel coefficient
}

grid_search = GridSearchCV(
    estimator=svm_rbf,
    param_grid=param_grid,
    cv=5,                     # 5-fold cross-validation
    scoring='accuracy',       
)

# Fit GridSearchCV to the scaled training data
grid_search.fit(X_train_processed, y_train)


# --- Report the best parameters and cross-validation score ---
best_params = grid_search.best_params_
best_score = grid_search.best_score_

print(f"Best Hyperparameters found: {best_params}") # Print as requested
print(f"Best Cross-Validation Accuracy: {best_score:.4f}") # Print as requested
````

**&#9989; (8 points) Task 4.3:** Evaluate the optimized SVM model:
- Train the model with the best parameters on the training data (2 points)
- Make predictions on the test set (1 point)
- Create a confusion matrix (1 point)
- Display the confusion matrix using a heatmap with Blues colormap (2 points)
- Calculate and print accuracy, precision, recall (2 points)


````python
best_svm_model = grid_search.best_estimator_
svm_predictions = best_svm_model.predict(X_test_processed)


# --- Create the confusion matrix ---
svm_cm = confusion_matrix(y_test, svm_predictions)

sns.heatmap(svm_cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Predicted 0', 'Predicted 1'],
            yticklabels=['Actual 0', 'Actual 1'])
plt.xlabel('Predicted Label')
plt.ylabel('Actual Label')
plt.title('SVM Confusion Matrix Heatmap')

svm_accuracy = accuracy_score(y_test, svm_predictions)
# Calculate precision and recall for the positive class (label 1)
svm_precision = precision_score(y_test, svm_predictions, pos_label=1)
svm_recall = recall_score(y_test, svm_predictions, pos_label=1)

print(f"\nOptimized SVM Model Evaluation Metrics:") # Print as requested
print(f"Accuracy: {svm_accuracy:.4f}")
print(f"Precision (for class 1): {svm_precision:.4f}")
print(f"Recall (for class 1): {svm_recall:.4f}")
````

**&#9989; Question 6.1 (4 points):**  A new disease has been detected in Sweden. Doctors have taken multiple measurements of patients and passed them along to you. You’ve tested various models to predict which patients have the new disease. The Swedish health experts have told you it is critical that they identify **all** of the patients with the new disease so they can put them into quarantine. They tell you that it doesn’t matter if your model accidentally flags some people as having the disease even when they don’t, as it’s much better to tell people who aren’t sick to quarantine than to tell a sick person they don’t need to quarantine.

Given this situation, should you choose the model that maximizes `recall` or the model that maximizes `precision`? Why?

*For this specific scenario involving disease detection in Sweden, the primary goal is to identify every single patient who has the disease to ensure they are quarantined. Therefore, the model selection should prioritize maximizing recall. Recall measures the proportion of true positive cases (actual sick patients) that are correctly identified by the model, and maximizing it inherently minimizes false negatives – instances where sick individuals are missed. While precision focuses on minimizing false positives (healthy individuals wrongly flagged as sick), the Swedish health experts have indicated that such errors are acceptable compared to the critical danger of missing even one sick person. Essentially, the goal here is to maximize true positives and minimize false negatives without care for false positives*

Precision is calculated as True Positives / (True Positives + False Positives), and maximizing it works to minimize the number of false positives – instances where legitimate applicants are wrongly flagged as fraudulent. Your boss explicitly stated that false accusations are unacceptable, even if it means some actual fraud cases (false negatives) are missed, making precision the key metric to optimize rather than recall, which focuses on minimizing missed cases.*
